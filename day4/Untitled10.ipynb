```python
import pandas as pd
import numpy as np

# Creating a DataFrame with NaN values
data = {'Name': ['Alice', 'Bob', np.nan, 'David'],
        'Age': [25, np.nan, 30, 22],
        'Salary': [50000, 60000, np.nan, 45000]}
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
print(df)
print("**** isna ***")
print(df.isna())
#print(df.isnull())
```

```python
df.dropna() 
```

```python
print(df)
```

```python
print(df)
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=2)
```
