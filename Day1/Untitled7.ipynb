```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_excel("salary_Data_1 (1).csv.xlsx")
```

```python
print(df.head())
```

```python
X = df[['YearsExperience']].values  # independent variable (must be 2D array)
Y = df['Salary'].values             # dependent variable (target)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
#
# initialization the linear regression model
regressor = LinearRegression()
```

```python
regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
```

```python
r2 = r2_score(Y_test, Y_pred)
```

```python
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
```

```python
print("\n--- Model Results ---")
print(f"R-squared (R\u00b2) Score (Closer to 1.0 is better): {r2:.4f}")
print(f"Root Mean Squared Error (RMSE): ${rmse:.2f}")
print(f"Model Intercept (\u03B8\u2080): ${regressor.intercept_:.2f}")
print(f"Model Coefficient (\u03B8\u2081 - Slope): ${regressor.coef_[0]:.2f}")
print("\nInterpretation: For every extra year of experience, the salary is predicted to increase by the coefficient value.")
```
